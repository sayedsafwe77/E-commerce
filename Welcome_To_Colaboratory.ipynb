```

#Q1--Map
from operator import add
import re
import time
start_time = time.time()

# Define constants
REGEX_PATTERN = "[^A-Za-z0-9]"



# Define the Log class
class Log:
    def __init__(self, code, title, hits, size):
        self.code = code
        self.title = title
        self.hits = int(hits)
        self.size = int(size)

# Define a function to create Log objects from strings
def create_log_obj(line):
    splits = line.split(" ")
    return Log(splits[0], splits[1], splits[2], splits[3])
def map_page_size(log):
    return log.size

# Define the reduce function to compute the min and max
def reduce_min_max(x, y):
    return min(x[0], y[0]), max(x[1], y[1])

# Define the map function to extract the page size and hits
def map_size_hits(log):
    return log.size, log.hits

# Define the reduce function to compute the most popular record with the largest page size
def reduce_max_hits(x, y):
    if x[0] == y[0]:
        return x if x[1] > y[1] else y
    else:
        return x if x[0] > y[0] else y

# Define a function to clean up the page title by removing non-alphanumeric characters
def clean_title(title):
    return re.sub(REGEX_PATTERN, "", title.lower().strip())

# Load the data
rdd_page = sc.textFile("pagecounts-20160101-000000_parsed.out").map(create_log_obj)

# Compute min, max, and average page size
page_size_min, page_size_max = rdd_page.map(map_page_size).map(lambda x: (x, x)).reduce(reduce_min_max)
page_size_sum = rdd_page.map(map_page_size).reduce(add)
page_size_count = rdd_page.count()
page_size_average = page_size_sum / page_size_count

# Print the output
with open("output 1 map.txt", "w") as f:
    f.write("Min page size: {}\n".format(page_size_min))
    f.write("Max page size: {}\n".format(page_size_max))
    f.write("Average page size: {}\n".format(page_size_average))
# Stop the Spark context
sc.stop()

end_time = time.time()
elapsed_time = end_time - start_time

print("Elapsed time:", elapsed_time, "seconds")
```

```
#Q1--Loops
import re
import time
start_time = time.time()

# Define constants
REGEX_PATTERN = "[^A-Za-z0-9]"

# Define the Log class
class Log:
    def __init__(self, code, title, hits, size):
        self.code = code
        self.title = title
        self.hits = int(hits)
        self.size = int(size)

# Initialize Spark context
sc = SparkContext("local", "Page Sizes")

# Load the data
with open("pagecounts-20160101-000000_parsed.out") as f:
    lines = f.readlines()

logs = []
for line in lines:
    splits = line.split(" ")
    log = Log(splits[0], splits[1], splits[2], splits[3])
    logs.append(log)

# Create RDD from logs
rdd_logs = sc.parallelize(logs)

# Compute min, max, and average page size
page_sizes = rdd_logs.map(lambda log: log.size)
page_size_min = page_sizes.reduce(lambda x, y: min(x, y))
page_size_max = page_sizes.reduce(lambda x, y: max(x, y))
page_size_sum = page_sizes.reduce(lambda x, y: x + y)
page_size_count = page_sizes.count()
page_size_average = page_size_sum / page_size_count

with open("output 1 Loops.txt", "w") as f:
    f.write("Min page size: {}\n".format(page_size_min))
    f.write("Max page size: {}\n".format(page_size_max))
    f.write("Average page size: {}\n".format(page_size_average))
# Stop the Spark context
sc.stop()

end_time = time.time()
elapsed_time = end_time - start_time

print("Elapsed time:", elapsed_time, "seconds")
```

```
#Q2---Map
# Define constants
ARTICLE_PREFIX = "The"
EN_PROJECT_PREFIX = "en"

# Define the Log class
class Log:
    def __init__(self, code, title, hits, size):
        self.code = code
        self.title = title
        self.hits = int(hits)
        self.size = int(size)

# Define a function to create Log objects from strings
def create_log_obj(line):
    splits = line.split(" ")
    return Log(splits[0], splits[1], splits[2], splits[3])

# Define the map function to extract the page title and project code
def map_title_project(log):
    splits = log.title.split("_")
    return splits[0], log.code

# Define the reduce function to count the number of page titles that start with "The"
def reduce_title_count(x, y):
    return x + y

# Define the reduce function to count the number of page titles that start with "The" but are not part of the English project
def reduce_non_en_count(x, y):
    return x + y

# Load the datafrom the input file
rdd_page = sc.textFile("pagecounts-20160101-000000_parsed.out").map(create_log_obj)

# Map the page titles and project codes
rdd_title_project = rdd_page.map(map_title_project)

# Filter the page titles that start with "The"
rdd_the_titles = rdd_title_project.filter(lambda x: x[0].startswith(ARTICLE_PREFIX))

# Count the number of page titles that start with "The"
the_title_count = rdd_the_titles.map(lambda x: 1).reduce(reduce_title_count)

# Filter the page titles that start with "The" but are not part of the English project
rdd_non_en_titles = rdd_the_titles.filter(lambda x: not x[1].startswith(EN_PROJECT_PREFIX))

# Count the number of page titles that start with "The" but are not part of the English project
non_en_title_count = rdd_non_en_titles.map(lambda x: 1).reduce(reduce_non_en_count)

# Print the output
with open("output 2 Map.txt", "w") as f:
    f.write("Number of page titles that start with 'The': {}\n".format(the_title_count))
    f.write("Number of page titles that start with 'The' but are not part of the English project: {}\n".format(non_en_title_count))

# Stop the Spark context
sc.stop()
```

```
#Q5--Map
import time
start_time = time.time()
# Load the input data into an RDD
data = sc.textFile("pagecounts-20160101-000000_parsed.out")


# Function to parse each line and extract the necessary fields
def parse_line(line):
    fields = line.split(" ")
    return fields[1], fields[2:]

# Combine data of pages with the same title
def combine_page_data():
    title_data = data.map(parse_line)  # Map each line to (title, page_data) pairs
    combined_data = title_data.groupByKey().flatMap(lambda x: [(x[0], list(x[1]))])  # Group by title and flatten the values

    return combined_data

# Combine page data using the map-reduce paradigm
combined_data = combine_page_data()

# Write the output to a file
with open("output5 map.txt", "w") as f:
    for title, data in combined_data.collect():
        f.write(f"Title: {title}\n")
        for page_data in data:
            f.write(f"Page data: {page_data}\n")
        f.write("\n")

# Stop the SparkContext
sc.stop()

end_time = time.time()
elapsed_time = end_time - start_time

print("Elapsed time:", elapsed_time, "seconds")
```

```
#Q5--Loops
import time
start_time = time.time()

data = sc.textFile("pagecounts-20160101-000000_parsed.out")

def parse_line(line):
    fields = line.split(" ")
    return fields[1], fields[2:]

# Combine data of pages with the same title
def combine_page_data():
    title_data = {}

    for line in data.toLocalIterator():
        title, page_data = parse_line(line)

        if title in title_data:
            title_data[title].append(page_data)
        else:
            title_data[title] = [page_data]

    combined_data = [(title, data) for title, data in title_data.items()]

    return combined_data

# Combine page data using Spark loops
combined_data = combine_page_data()

# Write the output to a file
with open("output 5 loops.txt", "w") as f:
    for title, data in combined_data:
        f.write(f"Title: {title}\n")
        for page_data in data:
            f.write(f"Page data: {page_data}\n")
        f.write("\n")




# Stop the SparkContext
sc.stop()

end_time = time.time()
elapsed_time = end_time - start_time

print("Elapsed time:", elapsed_time, "seconds")
```
